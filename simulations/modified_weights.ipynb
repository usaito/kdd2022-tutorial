```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def true_iw(w: np.ndarray) -> np.ndarray:
    return w 

def switch_iw(w: np.ndarray, lambda_: float) -> np.ndarray:
    return np.minimum(w, lambda_)

def shrinkage_iw(w: np.ndarray, lambda_: float) -> np.ndarray:
    return lambda_ * w / (w ** 2 + lambda_)

def smoothing_iw(w: np.ndarray, lambda_: float) -> np.ndarray:
    return w / (1. - lambda_ + lambda_ * iw)
```

```python
iw = np.arange(100)

plt.style.use('ggplot')
plt.subplots(figsize=(10, 6), tight_layout=True)
plt.plot(iw, true_iw(iw), label="True", linewidth=5)
plt.plot(iw, switch_iw(iw, lambda_=50), label="Switch ($\lambda=50$)", linewidth=4)
plt.plot(iw, shrinkage_iw(iw, lambda_=1e3), label="Shrinkage ($\lambda=10^3$)", linewidth=4)
plt.plot(iw, shrinkage_iw(iw, lambda_=1e5), label="Shrinkage ($\lambda=10^5$)", linewidth=4)
plt.plot(iw, smoothing_iw(iw, lambda_=1e-1), label="Smoothing ($\lambda=10^{-1}$)", linewidth=4)
plt.plot(iw, smoothing_iw(iw, lambda_=1e-2), label="Smoothing ($\lambda=10^{-2}$)", linewidth=4)

plt.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=0, fontsize=18)
plt.xlabel("True IW", fontdict=dict(size=25))
plt.ylabel("Modified IW", fontdict=dict(size=25))
plt.show()
```
